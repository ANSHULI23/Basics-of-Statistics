```python
#1. Read the dataframe, check null value if present then do the needful, check duplicate row , if present then do the needful.


import pandas as pd

# Step 1: Read the dataset
df = pd.read_csv('spotify.csv')  # Make sure the file is in the same folder or upload it in Colab

# Step 2: Check for null values
print("Null values in each column:")
print(df.isnull().sum())

# Step 3: Drop rows with null values (if any)
df = df.dropna()

# Step 4: Check for duplicate rows
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Step 5: Drop duplicate rows (if any)
df = df.drop_duplicates()

# Step 6: Display the cleaned dataframe
print("\nCleaned dataframe:")
print(df.head())
```

```python
#2. What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.


import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Popularity'], bins=30, kde=True)
plt.title('Distribution of Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
#3. Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Duration (ms)', y='Popularity', data=df)
plt.title('Popularity vs Duration')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.show()
```

```python
#4. Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('spotify.csv')

# Count the number of tracks for each artist
artist_track_counts = df['Artist'].value_counts()

# Display the artist with the highest number of tracks
print("Artist with the highest number of tracks:")
print(artist_track_counts.idxmax(), "with", artist_track_counts.max(), "tracks")

# Plot the count of tracks for each artist
plt.figure(figsize=(12, 8))
sns.countplot(y='Artist', data=df, order=artist_track_counts.index)
plt.title('Number of Tracks per Artist')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.show()
```

```python
#5. What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.

# Sort the dataframe by popularity (ascending order) and select the top 5
least_popular_tracks = df.sort_values(by='Popularity').head(5)

# Display the artist and track name for the top 5 least popular tracks
print("Top 5 least popular tracks:")
print(least_popular_tracks[['Artist', 'Track Name', 'Popularity']])
```

```python
#6. Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist.

# Group by artist and calculate average popularity
artist_avg_popularity = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False)

# Select the top 5 most popular artists
top_5_artists = artist_avg_popularity.head(5)

# Display the artist with the highest average popularity
print("Artist with the highest average popularity among the top 5:")
print(top_5_artists.idxmax(), "with an average popularity of", top_5_artists.max())

# Display the average popularity for each of the top 5 artists
print("\nAverage popularity for the top 5 artists:")
print(top_5_artists)
```

```python
#7.  For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.


# Get the top 5 most popular artists
top_5_artists = df.groupby('Artist')['Popularity'].mean().nlargest(5).index

# Find the most popular track for each of the top 5 artists
most_popular_tracks = df[df['Artist'].isin(top_5_artists)].sort_values(by='Popularity', ascending=False).drop_duplicates('Artist')

# Display the most popular track for each artist
print("Most popular tracks for the top 5 artists:")
print(most_popular_tracks[['Artist', 'Track Name', 'Popularity']])
```

```python
#8. Visualize relationships between multiple numerical variables simultaneously using a pair plot.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('spotify.csv')

# Select numerical columns for the pair plot
numerical_columns = ['Popularity', 'Duration (ms)']  # Use valid numerical columns from your dataset

# Drop rows with null values in the selected columns
df = df.dropna(subset=numerical_columns)

# Create a pair plot
sns.pairplot(df[numerical_columns])
plt.show()
```

```python
#9. Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.

# Create a box plot or violin plot for duration across artists
plt.figure(figsize=(12, 8))
sns.boxplot(x='Duration (ms)', y='Artist', data=df, order=df['Artist'].value_counts().index[:10])  # Top 10 artists for clarity
plt.title('Duration of Tracks by Artist (Box Plot)')
plt.xlabel('Duration (ms)')
plt.ylabel('Artist')
plt.show()

# Alternatively, use a violin plot
plt.figure(figsize=(12, 8))
sns.violinplot(x='Duration (ms)', y='Artist', data=df, order=df['Artist'].value_counts().index[:10])
plt.title('Duration of Tracks by Artist (Violin Plot)')
plt.xlabel('Duration (ms)')
plt.ylabel('Artist')
plt.show()
```

```python
#10. How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.

# Create a violin plot for popularity across artists
plt.figure(figsize=(12, 8))
sns.violinplot(x='Popularity', y='Artist', data=df, order=df['Artist'].value_counts().index[:10])  # Top 10 artists for clarity
plt.title('Popularity Distribution by Artist (Violin Plot)')
plt.xlabel('Popularity')
plt.ylabel('Artist')
plt.show()

# Alternatively, use a swarm plot
plt.figure(figsize=(12, 8))
sns.swarmplot(x='Popularity', y='Artist', data=df, order=df['Artist'].value_counts().index[:10])
plt.title('Popularity Distribution by Artist (Swarm Plot)')
plt.xlabel('Popularity')
plt.ylabel('Artist')
plt.show()
```
